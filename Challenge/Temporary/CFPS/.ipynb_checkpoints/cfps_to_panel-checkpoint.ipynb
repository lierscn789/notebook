```python
import os
import re
import pandas as pd
```

```python
# 参数设置
DATA_FILE_PATH = r'E:\datahouse\rawdata\microdata\final2'
```

```python
# 导入2012年个体数据
raw_individual_2012_data = pd.read_stata(os.path.join(DATA_FILE_PATH, '2012conf.dta'), convert_categoricals = False)
raw_individual_2012_data.iloc[0:10,]
```

```python
# 计算年龄
def to_age(birth_date,report_age,invest_year):
    if birth_date > 0:
        age = invest_year - birth_date
    else:
        age = -999
    if report_age is not None:
        if int(report_age) != int(age):
            if report_age >= 0:
                age = report_age
    
    return age
```

```python
raw_individual_2012_grouped_data = raw_individual_2012_data.groupby('fid12')
variables = list(raw_individual_2012_data.columns)
family_memeber_keywords = ['f','m','s']
family_memeber_keywords.extend([''.join(['c',str(i)]) for i in range(1,11)])

# 用来判别是否已经计入数据中
fid_set = set()
filtered_data_list = []
for fid, grouped_data in raw_individual_2012_grouped_data:
    grouped_data.index = range(grouped_data.shape[0])
    
    for ind in grouped_data.index:
        # 处理年龄，活着的才计入，无信息的年龄设定-999
        if grouped_data.loc[ind,'alive_a_p']==1:
            age = to_age(grouped_data.loc[ind,'tb1y_a_p'],grouped_data.loc[ind,'tb1b_a_p'],2012)
            
            if age < 0:
                print(grouped_data.loc[ind,'pid'],grouped_data.loc[ind,'tb1b_a_p'],grouped_data.loc[ind,'tb1y_a_p'],age)
            fid_set.add(grouped_data.loc[ind,'pid'])
            filtered_data_list.append([grouped_data.loc[ind,'pid'],grouped_data.loc[ind,'fid12'],age,grouped_data.loc[ind,'familysize']])

        for key_word in family_memeber_keywords:
            if grouped_data.loc[ind,''.join(['alive_a_',key_word])] == 1:
                age = to_age(grouped_data.loc[ind,''.join(['tb1y_a_',key_word])],grouped_data.loc[ind,''.join(['tb1b_a_',key_word])],2012)
            
```
