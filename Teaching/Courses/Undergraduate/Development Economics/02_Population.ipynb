## 影响人口增长的原因

### 马尔萨斯人口理论

1798年, 32岁的马尔萨斯匿名发表了《人口原理》第一版。该书第二版于1803年署名出版, 由一本小册子变成篇幅大4倍的巨著, 被誉为人口科学的开山之作。《人口原理》第一版的书名为《人口原理:人口对社会未来进步的影响,兼评葛德文先生、孔多塞先生和其他著述家的推测》;第二版的书名为《人口原理对于人类幸福之过去及现在之诸影响的考察。附考察将来关于消除或缓和由人口所生的弊害的研究》。

<br>

#### 两个公理

- 食物为人类生存所必需
- 两性间的情欲是必然的，且几乎会保持现状

<br>

#### 两个级数

- 人口在无妨碍时以几何级数增加,而生活资料只以算术级数增加。
- 人口增殖力和土地生产力增殖之间不平衡,前者要大于后者。

<br>

#### 三个命题

- 人口增加必然受到生活资料的限制
- 只要生活资料增长,人口一定会增长,除非受到某种抑制
- 某种抑制使得现实的人口与生活资料相平衡

<br>

#### 两种抑制

- 预防抑制—— 一种通过限制结婚等手段控制人口增长的方式，存在于近代欧洲
- 积极抑制—— 一种是人口无节制的增长直至日益增长的贫困导致死亡率上升, 存在于非近代世界和欧洲以外的地区

<br>

#### 马尔萨斯人口理论的基本思想

马尔萨斯认为人口增长有永远快于生活资料增长的趋势, 如果不加以限制, 人口总是按几何级数增加而生活资料却只能按算术级数增长,当人口扩张到生活资料仅能维持生存的极限时, 两者之间的不可调和的矛盾必然爆发, 此时唯有预防性的抑制(如不婚、晚婚、不生育等) 和积极的抑制(如贫困、饥谨、罪恶、灾荒和战争等)才能遏止过度的人口增长, 最终使人口数量与能够承载其生存的生活资料的数量保持平衡。

<br>

#### 马尔萨斯人口陷阱

“人口陷阱” , 即“马尔萨斯均衡陷阱”,指在巨大的人口增殖力下,人类始终处于贫困线上,无法摆脱的一种人口与经济的恶性循环。

<br>

#### 马尔萨斯人口理论评价

> “两个级数”在今天看来过于简单, 但它开创性地探讨了物质资料生产与人类自身再生产相协调的人口规律问题, 并对人口变动进行了科学的定量解读, 影响深广。

> 坚决反对以罪恶和贫困等手段来消灭过剩人口, 并认为这一类主张“是一种一刹那也不能容许的思想” 。相反, “罪恶和贫困, 也只有这些, 是我的主要斗争对象。”

> 马尔萨斯把有效抑制人口增长、改善人民生活的社会途径归纳为: 在一种充满就业机会、多样性选择的社会条件下, 将会最大限度地激发人们的个人努力, 并在有效的法律保护下激励人们遵从一种更加理性的生活方式。

<br>

**讨论：生育率下降**

当今世界人口问题已成为一个突出的问题。全球许多国家都陷入了“低生育率陷阱”，中国目前的总生育率约为1.6，远低于世界公认的维持人口正常替代水平的2.1，因此中国也面临着跌入“低生育率陷阱”的可能性。

### 新家庭经济学理论

新家庭经济学在分析人们的生育行为时,引入微观经济学的消费者行为理论。它假设在是否要生孩子这个问题上,一对夫妇依据家庭效用最大化的合理准则,即以人的偏好一定为前提,在家庭收入有限这个约束条件下,来实现家庭消费最优(效用最大)这个目标函数。

新家庭经济学使用的消费函数其构成不是传统意义上的商品,而是“新消费者论”强调的各种抽象物品。比如,“人们购买汽车不是为了汽车本身,而是为了得到以汽车为物质基础的服务和满足”。

#### 孩子对父母的效用

根据哈维莱宾斯坦的划分,孩子对父母的效用有三种

- 给父母带来精神上的满足;即直接得自孩子自身的效用,这时孩子是品种消费品
- 作为一种生产性物品,他可能很小就参加劳动,为家庭赚取收入
- 作为一种保障的资源,以期孩子为父母晚年提供生活保障

### 人口转型与人口红利

#### 人口转型

人口转型指的是人口生育率、死亡率和人口结构的变化。人口转型现象大致可描述为, 首先是死亡率的下降, 接着是生育率的下降, 人口规模的增长则是先加速而后减速, 最后是人口老龄化的阶段。

<br>

#### 人口转型的三个阶段

- 第一阶段为高出生率、高死亡率、低人口增长率
- 第二阶段为高出生率、低死亡率、高人口增长率
- 第三阶段为低出生率、低死亡率、低人口增长率

<br>

#### 人口机会窗口

人口转型过程中，在第二阶段向第三阶段的过渡时期，此时， 过去的高出生率使得劳动年龄人口还在持续增长，之前的低出生率又迫使少儿抚养比迅速降低，因此劳动力供给充沛，劳动年龄人口规模巨大，总抚养负担比较轻，中国学者把这个阶段称为人口机会窗口。

<br>

#### 人口红利

当一个国家经历特殊的人口发展阶段时———劳动年龄人口增加和抚养比下降， 即使缺少技术创新和改革， 也有机会获得快速的经济增长。而这种特殊的人口结构对经济增长的贡献， 我们称之为“人口红利”。

<br>

#### 人口机会窗口与人口红利

人口机会窗口开启并不意味着必然存在人口红利，二者之间并不具有等价关系。只有当经济社会条件配合人口机会窗口开启，才能将潜在人口机会转变为现实的经济红利。

<br>

#### 人口红利的两个主要特征

- 持续增加的劳动年龄人口将为经济增长提供充足的劳动力供给
- 人口抚养比下降， 消费将减少而储蓄会增加， 这将有利于资本形成， 从而为经济增长提供充足的资本供给

<br>

#### 人口红利的衡量

人口抚养比，即劳动人口和被抚养人口直接的比例。
