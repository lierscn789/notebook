# 自然资源、环境和可持续发展

- 荷兰病与“资源的诅咒”

## 荷兰病与“资源的诅咒”

### 荷兰病

**荷兰病**

荷兰病指的是一个国家随着资源的开发而发生制造业萎缩、国内物价上涨以及本币升值的现象。

**荷兰病的机制**

- 资源的开发导致劳动力和资本向资源部门流动，从而迟滞制造业的增长；同时，资源开发带来的财富的增加，从而提高本国不可贸易品的价格
- 资源的大量出口导致国际收支的改善，本币升值

**荷兰病的例子**

- 苏联的解体
- 巴西

**避免荷兰病的途径**

- 把资源开发带来的财富转变为对未来生产力的投资
- 基础设施和机器设备这样的制造业赖以生存的物质资本积累
- 人力资本的积累

**中国房地产业扩张与“荷兰病”**

中国房地产行业的扩张通过资源的重新配置和财富效应，在一定程度上推动了物价上涨和人民币实际升值，并对制造业的出口产生负面影响其在本质上也是一种“荷兰病”。

### 资源的诅咒

“资源的诅咒”的含义要宽泛得多，也模糊得多。它既可以指一个国家（地区）的人均收入没有因为资源开发而提高的现象，也可以指一个资源丰富的国家（地区）比一个资源贫乏的国家（地区）增长更慢的现象。

**发生资源的诅咒的经济原因**

- 它可能是荷兰病的后果之一
- 资源的开发成本相对较低，因而吸引大量资本和劳动力，受限于边际报酬递减的限制；相反，资源比较贫乏的国家和地区，更注重人力资本和技术进步的投资，经济增长具有可持续性

**发生资源的诅咒的政治原因**

- 资源开发容易滋生腐败
- 丰富的自然资源容易成为寡头政治的温床

## 环境污染与经济发展

**环境库兹涅茨曲线**

它指环境污染与经济增长的长期关系呈倒U形。其含义是:当一国经济发展水平较低时环境污染较轻,但其恶化程度随经济增长而加剧,当该国经济发展达到一定水平后,环境质量会逐渐改善。

> 经济结构解释
> 它是规模效应和经济结构自然演进双重作用的结果。
> - 随着人均收入提高,经济规模变得越来越大,需要资源投入越来越多,而产出的提高意味着经济活动副产品———污染的增长,从而使得环境状况恶化,这就是规模效应
> - 当经济发展到更高水平时,产业结构升级,能源密集型为主的重工业向服务业和技术密集型产业转移,污染减少,这就是结构效应

## 可持续发展

它指满足当前需要但又不削弱后代子孙后代生存利益。

**可持续发展的内涵认知**

- 人类向自然的索取能够被人类向自然的回馈相平衡
- 人类对于当代的努力能够同对后代的贡献相平衡
- 人类为本区域发展的思考能够同时考虑到其它区域乃至全球利益
