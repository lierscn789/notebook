## 工业化与结构转变

- 自从工业革命使英国的工业生产在19世纪上半叶提高了400%，发展的概念和工业化进程就通常被认为是同义的。
- 从那时起直到现在，发展的最主要的标准就是通过工业化使人均收入获得大幅度的提高。
- 对于发展中国家来说,工业化是经济发展过程的一个重要阶段，也是产业结构迅速转变的一个重要时期。

**配第-克拉克定理**

随着经济的发展、人均国民收入水平的提高，劳动力首先由第一产业向第二产业转移；当人均国民收入进一步提高时，劳动力便向第三产业转移。这一产业结构演变的基本趋势被称为配第-克拉克定理。

> **劳动力转移的原因**
> 克拉克认为，劳动力在产业间的转移，其根本原因是由经济发展中各产业间出现收入（附加值）的相对差异造成的。

## 城市化

城市化是由农村向城市转变的过程，它既包括人的转移和生产方式的改变，也包括地理意义上的转换和生活方式的改变。

## 剩余劳动力和二元经济结构

**剩余劳动力理论**

刘易斯（1954）年提出关于传统农业的剩余劳动力理论。他认为，传统农业里存在经常性的劳动力过剩，这些过剩人口获得生存工资，却不增加农业产出，所以存在所谓的“劳动力的无限供给”。

**二元经济结构理论**

刘易斯的理论经由兰尼斯和费景汗（1961）进一步发展为二元经济结构理论。即一个发展中经济需经历两个阶段：第一阶段存在剩余劳动力，城乡市场不统一，即形成“二元结构”；第二个阶段剩余劳动力完全被工业吸收，城乡结构统一，进入现代化经济阶段。

**剩余劳动力**

按照刘易斯的最初定义，剩余劳动力指的是边际产出为零的劳动力。实际上，刘易斯的剩余劳动力是对人口过剩的一个定义。

**制度工资**

所谓制度工资，就是让一个劳动力得以存活的工资，因此也称为生存工资。

## 托达罗模型

**问题**

如何城市失业与乡村人口向城市涌入两者并存的现象？

> 传统人口流动模型
> 基本假设工业部门可以实现完全就业

托达罗引入了以下概念来阐述他的模型

- 非正规部门
- 就业概率
- 预期收入

托达罗的模型反映了更真实的人口流动：人们在由农村进入城市的过程中，往往难以一步到位、进入正规部门就业，而是更多的先被非正规部门吸纳。
