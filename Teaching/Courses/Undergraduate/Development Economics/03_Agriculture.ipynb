## 传统农业

### 传统农业的典型特征

不使用现代要素的农业生产都被视为传统农业的范畴，传统农业的特征是“贫穷但有效”（舒尔茨，1964）。

- 传统农业的“贫穷”来自于它具有的较低的生产力水平
    - 不使用良种、化肥、杀虫剂、机械等现代农业生产要素
    - 依赖三大传统要素，即土地、人力和畜力进行生产，面临要素边际报酬递减的限制


- 传统农业的“有效”来自于它的资源配置效率—— 帕累托最优
    - 要素边际产出等于要素价格
    - 要素市场出清

## 传统土地关系和土地改革

### 中国土地制度的历史回顾

- 井田制
    - 井田制是记载在《孟子》、《周礼》等儒家经典中的中国最古老的土地制度，即西周时期的土地制度
    - 井田是在小城圈之内，在封建制度下的小规模农业经济
    - “井田者，九百亩，公田居一。私田稼不善，则非吏；公田稼不善，则非民。” —— 春秋
    - “井”是灌溉单位，八家公用一井。一井的土地，即中间一格100亩属“公田”外，其余每家100亩，共九格
    - 公田的收成归封建主，私田则归农户自享


- 秦代的土地私有制
    - 秦始皇三十一年时，人民可想政府呈报自己拥有之田地。即所谓的“自实田”之意
    - “改帝王之制，除井田，民得买卖” —— 汉书
    
    
- 汉承秦制
    - 汉代高祖时行轻税政策；文景帝至西汉末年，田税一直是三十税一
    - 汉代人头税：算赋（15-26）和口赋（7-14）


- 东汉与三国的屯田制度
    - 汉武帝通西域时，军队是农兵两兼，移民殖边
    - 东汉初屯田颇有成效，使国有余量，导致田租减至三十税一
    - 曹操屯田统一中原；耕者无其地
    

- 西晋的土地制度
    - 品官占田制 —— 每一官员可占有之田亩数量受到限制
    - 户调制 —— 将土地分配给农民，并规定一个确定的土地占有数量，有名无实


- 北魏的均田制度
    - 即不论贫富，一律征收赋税
    - 其用意并不在田亩制绝对均给，只求富者有一限度，贫者亦有最低之水平


- 隋唐的土地制度
    - 隋朝颁布“均天下之田”，推行均田制的土地制度
    - 唐初的租佣调制
        - 18岁以上者授田一顷，其中80亩为口分田，年老时还给官府；20亩为永业田，可传给子孙，官府不再收回
        - 凡经营工商者，则永业田、口分田，各减半给之
        - 凡道士给田30亩，女冠20亩，僧20亩，尼20亩
        - 官员授永业田，亲王100顷，正一品官60顷...
    - 两税制
        - 两税制是认田不认人


- 宋代的方田制
    - 千步丈量法
    - 务求租税平均

### 农民的类型

- 地主
- 自耕农
- 佃农
- 雇农

**佃农**

宋代以后,个体小农独立生产能力逐渐增强,至明清趋于成熟(龙登高,1997)。近世的佃户,已经获得人身自由,可以自由迁徙,自由占有土地与种植,自由支配劳动时间与劳动产品(方行,2004)。往往拥有自己的主要生产资料与工具,如耕牛、农具、种子,还有家庭代际传承的生产技术,资本亦往往由自家筹集,或通过家族互济。家庭劳动力男女老幼搭配,农忙时则求助于短工市场或邻里亲友互助,农闲时外出从事副业。土地租佃通常是契约关系,近世中国佃农对地主的经济依赖性越来越弱,在经济上地主往往并没有主导权(赵冈,2003) 。佃农既可以雇入农民帮工,也可以短期出卖自己的劳动力;既可以租入土地,定额租制独立经营,又可以与地主分成,还可以转佃与转租。可见,与自耕农、地主相比,佃农具有更大的弹性与发展前景。

**雇农**

毛泽东在《兴国调查》中写道“:雇农没有老婆的占百分之九十九,是农村中最苦的一个阶级。地主富农不但人人有老婆,一人几个老婆的也有。中农百分之九十有老婆,百分之十没有。贫农百分之七十有老婆,百分之三十没有。手工工人百分之七十有老婆,百分之三十没有。游民百分之十有老婆,百分之九十没有,也比雇农中有老婆的多些,只有雇农才是百分之九十九无老婆。”

### 土地改革

- 进行土地改革的主要动机通常是政治上的，而非经济上的。
- 土地改革对农业生产率的影响，取决于改革的是何种体制以及改革所采用的措施
- 世界范围内的经验表明，较为平均的土地分配有助于经济增长

## 现代农业

- 现代农业的主要特征是“现代要素”投入，例如机械，生物科技等
- 现代农业制度和组织形式，例如适当的土地制度，农村基础设施，金融服务，技术推广和市场发展

### 农业现代化

一个国家实现农业现代化, 究竟采用哪种起步方式, 一般地说, 主要是由当时的土地、劳动力和工业化水平决定的。人少地多的国家, 首先从生产工具上进行改革, 发展机械化以节约劳动力;人多地少的国家,则多投入劳动力, 充分利用土地以提高单产水平。

农业现代化可分为美国模式、日本模式和西欧模式：美国模式是指在人少地多、劳动力资源相对稀缺的国家, 凭借发达的现代工业和低价能源优势, 大力发展农用机械, 以机械取代人力和畜力, 通过扩大经营面积提高农作物的总产量, 以提高劳动生产率为主的农业现代化模式；日本农业现代化突出特点就是把科技进步放在重要位置, 通过改良农作物品种, 加强农田水利设施建设, 发展农用工业, 提高化肥与农药施用水平, 致力于提高单位面积产量, 以提高土地生产率为主要目标；西欧在农业现代化的过程中既重视现代工业装备农业, 又重视现代科学技术的普及与推广, 机械技术与生物技术并进, 以提高劳动生产率和土地生产率并重为主要目标。把农业生产技术现代化和农业生产手段现代化放在同等重要的地位,实现了农业机械化、电气化、水利化、园林化, 同时在土地产出率和劳动生产率两方面得到提高, 典型的国家有英国、法国、德国等。

## 农业在经济发展中的作用

### 农业的特点

- 对于早期社会（国家）而言，农业部门所使用的的人力远远超过其他所有行业和部门的总和
- 农业活动已经存在数千年
- 土地在农业活动中的极端重要性
- 农业是唯一生产食品的部门

### 农业在经济发展中的作用

- 贫困国家的大多数人都是靠土地谋生的
- 发展中国家的农民不仅要生产足够的粮食养活自己，还要生产足够的粮食来供养城市人口
- 农业在为工业和其他现代部门提供生产要素投入——主要是劳动力方面起着十分重要的作用
- 农业部门也能成为现代经济增长所需资本的一个主要来源
- 发展中国家的农业人口，至少在某些情况下是现代城市部门产品的重要市场

## 中国农业的发展

### 中国的小农体系

小农体系指的是一种以小农耕作作为主导的农业生产组织形式。

小农经济的特点：

- 以家庭为单位
- 完全或主要依靠家庭成员的劳动
- 独立经营小规模农业，以满足自身消费需要为主

### 改革之前农业的表现

- 中国农业总产量有所增加，但生产率和收入都停滞不前
- 农业没有结构性地转向一个生产率更高、更高效的农业经济
- 农业部门的农民无法流动到其他部门
- 土地改革：土地合作社——人民公社
- 改革之前的物价政策没有起到鼓励生产和有效分配商品服务的作用
- 户籍制度人为地牵制了该时期农业结构转变的步伐

### 改革时期农业的表现

- 中国农业的产出有了飞速地增长
- 中国农业生产逐步不再以粮食为主，家禽业和渔业部门的增长速度已经超过了粮食部门
- 农业的相对重要性越来越不显著
- 中国农产品出口由1990年的不到100亿美元，增长到2003年的212:4亿美元
- 在20世纪90年代，中国农业全要素生产率以大约每年2%的速度持续增长——领先于国际水平
- 物价政策的变化，农产品价格市场化带来的巨大影响
- 始于1979年的中国农业经济改革以家庭联产承包责任制为基础
- 农村公共部门的改善
