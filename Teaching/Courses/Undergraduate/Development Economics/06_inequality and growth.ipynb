## 公平与不平等

**公平的内涵**

- 由于不同国家和文化传统的影响，公平有着不同的内涵
- 在大多数人的日常理解中，公平有时说的是平均或均等

**公平涉及的三个方面**

- 机会均等
- 过程公平
- 结果均等

世界银行认为：政策的目标是促进机会均等、消除极端贫困、但不在于实现结果完全均等

**促进公平的机制**

- 机会公平的要以在于投资于人的建设——教育、公共卫生和医疗服务、社会保障
- 过程公平的核心在于权利保护
- 通过一定的公共治理机制和公共政策消除极端贫困

## 不平等的度量

**衡量收入分配不平等的原则**

- 匿名原则—— 任意两个人的收入对换，不影响全社会的不平等程度
- 人口原则—— 人口总数与不平等程度无关
- 相对收入原则—— 收入分配不平等程度与收入水平无关，只有收入份额才有意义
- 累退原则—— 累退的收入转移使收入分配状况恶化

### 洛伦茨曲线

- 如果洛伦茨曲线和45度线重合，则任何比例的人口的收入占总收入的比例等于其人口比例，此时，收入分配完全均等化。
- 只要存在任何的收入分配不均，洛伦茨曲线必然在45度线以下。
- 曲线离45度线越远，则收入分布就越不平等。
- 洛伦茨曲线不但可以用来表示收入的分布，也可以用来表示财富的分布，衡量一个国家（地区）人均或家庭财富的不平等程度。
- 如果两个国家的洛伦茨曲线交叉，则我们无法判断哪个国家的收入分配更不平等。

### 基尼系数

- 基尼系数是洛伦茨曲线和45度线之间所围成新月形的面积占45度线下方与横轴围成的三角形的面积的比例。
- 基尼系数的取值范围在0和1之间.值越大，则收入分配越不平等。
- 当洛伦茨曲线是45度线时，基尼系数为0；当只有一个人拥有全部收入，其他所有的人的收入都是0的时候，基尼系数为1。
- 基尼系数可以简单直观反映收入分配的整体情况，但是对处于极端位置的收入人群不太敏感。
- 一个国家在公布基尼系数的时候，一般也会同时公布最富有的$20\%$的人口占全部收入的比例和最贫穷的$20\%$的人口占全部收入的比例。


### 不平等的程度

- 一般认为，当一个国家的基尼系数超过0.4的时候，这个国家的收入分配就是比较不平等的。
- 超过0.5，则这个国家就是非常不平等的。
- 目前，北欧国家是世界上收入分配最平等的国家，那里的基尼系数在0.3以下。
- 而最不平等的国家主要在非洲和拉丁美洲，它们的基尼系数超过0.55。
- 亚洲国家（地区）处于两者之间，通常在0.4 - 0.5之间。

## 贫困

<img src="./pic/ineq05.png" width = 60% height = 60% />

**贫困标准**

通常认为，贫困意味着“基本需要”得不到满足。

- 直接法—— 直接考察人们的消费组合是否满足对食品、服装、住房、交通等基本需要
- 收入法—— 所谓贫困是指收入不足以满足规定的最低需要

**绝对贫困**

所谓绝对贫困，就是那些没有能力获取满足基本需要资源的人群，凡生活在某一特定的最低的实际收入水平之下的人都计算在内。

**多维贫困**

贫困是一种复杂而综合的社会现象。

**贫困测度**

- 贫困广度指数—— 生活在贫困线以下的人口在总人口中所占的比重
- 人类贫困指数HPI —— 它由寿命（Longevity）、读写能力（Literacy）和生活水平（Living Standard）三个指标构成

**贫困的原因**

累积因果循环论

**库兹涅茨曲线**

库兹涅茨（Simon Kuznets）在上世纪五十年代研究英国的历史数据的时候，发现英国的收入不平等和人均收入之间呈现一个倒U型关系，即当收入较低的时候，不平等程度随着收入的增加和提高，当收入超过一定水平之后，不平等程度随着收入的增加而下降（Kuznets, 1955）。后人把这个发现称为“库兹涅茨曲线”。

> 市场经济条件下人口分化的一个重要表现是财富的货币化进程。当经济不甚发达的时候，个人财富往往仅仅是消费来源而已，而不能带来更多的财富；随着市场的扩展，财富（既包括现金，也包括土地、房产等非现金财富）的变现和互换变得更加灵活和多样，因此比较容易带来新的财富，从而扩大个人之间差距。

> 随着收入水平的提高，资产的影响越来越小，人力资本的作用越来越大，而人力资本的积累较资本的积累来得容易一些，因而其差距比资产的差距要小很多，因此不平等会下降。

**不平等影响经济增长的途径**

- 经济机制
    - 不平等限制市场大小
    - 不平等提高生育率
    - 不平等降低全社会人力资本积累速度
    - 不平等阻碍企业的发展

- 政治机制
    - 不平等导致过度分配和动态分配
    - 社会和谐
    - 避免资源与生产力之间的错配

**中等收入陷阱**

一个国家人均收入达到世界中等水平后，由于不能顺利实现经济发展方式的转变，导致增长动力不足，最终出现经济停滞的一种状态。

> 达到中等收入之后，单纯的资本积累必然面临边际报酬递减的制约

> 低技术产业的发展达到饱和状态，其他低收入国家加入竞争行列，挤压本国低技术产品的国际空间

> 技术进步和人力资本的积累没有跟上，从而使得产业升级后继乏力

> 总之，当一个国家进入中等收入后，以往单靠扩大外延的增长模式已经不适用了，需要寻找新的、以技术进步为主导的增长模式，因而科技投入和人力资本的积累变得非常重要。
