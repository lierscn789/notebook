## 贸易与发展

**绝对利益理论**

亚当$\cdot$斯密认为，如果两个国家在某些特定商品的生产上各自占有绝对的优势，按照分工原则，两国各自从事不同商品的生产将是最有效的，两国之间的贸易将会给双方都带来利益。

**比较利益理论**

李嘉图证明，即使一个国家在任何一种产品的生产商都没有绝对优势，但只要生产各种商品的相对成本和贸易伙伴并不完全一样，就一定会在某种商品的生产上占有比较优势。各国都应专业化生产本国相对成本较低的商品，这样就能通过国际贸易获得比较利益。

**要素禀赋理论**

要素禀赋理论认为，贸易的产生不是因为不同国家具有不同的劳动生产率，而是因为不同国家具有不同的要素禀赋。

- 资本相对丰富的国家，资本要素相对便宜，该国就在密集使用资本要素生产的商品上成本较低，因而具有比较优势
- 劳动相对丰富的国家，劳动要素相对便宜，该国就在密集使用劳动要素生产的商品上成本较低，因而具有比较优势

**列昂惕夫之谜**

列昂惕夫(Wassily Leontief)在20世纪50年代对美国进行了投入产出分析，发现了一个不符合要素禀赋论的现象：美国作为世界上资本最丰富的国家，按理应主要出口资本密集型产品、进口劳动密集型产品，而实际情况却与此相反，美国出口商品的资本—劳动比率，低于进口商品的资本—劳动比率。这被称为列昂惕夫之谜。

> 列昂惕夫之谜的解释
> - 可能是美国人对资本密集型产品的偏好特别高，以至于本国产品还不能满足国内需求。
> - 所谓劳动密集型产品也不能一概而论。美国生产的劳动密集型产品主要是人力资本密集型产品，其中凝结着高水平的知识和技术，与一般发展中国家生产的劳动密集型产品不可相提并论。这也是被广泛接受的一种解释。
> - 现实中某种产品到底是哪种生产要素密集型产品并不像初看上去那么一清二楚。比如同样是粮食，美国生产的粮食就是资本密集型的，而泰国生产的就是劳动密集型的。
> - 要素禀赋理论可能不适合解释自然资源基础上的出口。比如，出口石油的国家要开采石油，必须先期投入大量的资金，这就使石油这种产品显现为资本密集型，但出口石油的国家原本并非资本丰富的国家。

### 开放条件下一个国家发展的四个阶段

- 第一阶段，不仅在农产品上具有比较优势，而且会专业化生产农产品，资本存量会减少到零，工业品完全依赖进口
- 第二阶段，仍然在农产品上具有比较优势，但不是专业化生产农产品，同时也生产工业品，尽管可能也需要进口工业品
- 第三阶段，同时生产农产品和工业品，但出口工业品，同时可能进口农产品
- 第四阶段，不再生产农产品，专业化生产工业品

### 保护幼稚产业政策

- 压低农产品价格

- 进口替代政策—— 即对工业进口品征收高额关税，提高工业品的国内价格

- 进口替代的手段
    - 保护性关税
    - 进口配额
    - 补贴
    - 汇率管理

这些贸易壁垒提高了进口产品的价格，使国内生产商可以给产品标高价以弥补较高的生产成本并有利可图。但该政策是以牺牲消费者的福利为代价。

**进口替代的结果**

- 政府选择了保护不适当的行业，或者不愿意将保护手段从政治关系较好的行业中撤销。
- 一旦国内市场达到饱和，增长通常会停顿下来。企业无法利用经济规模的优势。
- 进口替代限制了新的技术和理念
- 面临由持续增长的贸易赤字带来的国际收支问题
- 激发寻租行为

### 出口对经济增长的贡献

- 打破国内市场狭小的限制

- 从比较优势的角度，出口有利于发展中国家更好地利用本国资源

- 提供就业

- 出口是一种储蓄行为

**出口导向**

从20世纪80年代末期，许多国家都将它们的贸易政策从进口替代转移到了更为外向型的政策上，以使企业能够应对激励的国际竞争。与进口替代政策最主要的区别在于，该政策是在促进国际竞争而非保护企业免遭竞争，其宗旨在于通过激发投资、提高产量、促进学习与引进新技术来促进经济增长。比较典型的模式是，在初级阶段，企业大多生产例如纺织品、服装、电器等相对简单的劳动密集产品进行出口。然后，随着工人们不断掌握新的技巧和技术，企业开始转向生产较为复杂的产品，一国的比较优势逐渐转向生产更为先进的电子设备、更高端的衣服以及耐用消费品。

## 外资利用和发展

**外商直接投资的利弊**

- 外国资本同本国资本相比要不稳定得多
- 外国资本对于就业的贡献，要视乎其所在的产业
- FDI的投资往往更有效率，能够帮助提高生产的专业化
- FDI也使接受国获得了进入世界市场的机会
- FDI的溢出效应，例如科技、技术和观念的转移，训练工人和管理人员
- FDI也能产生负的溢出效应，例如环境破坏，空气和水的污染
- 也许对于FDI最大的忧虑是东道国失去对于相关业务的控制

### 中国的国际贸易、外资利用和经济发展

**改革前的贸易体制**

- 1978年前，进出口计划是由外贸部所辖或控制的少数几个外贸公司负责执行的
- 中国的贸易结构与其比较优势关系不大，甚至一直到80年代初，资本密集型产品包括精炼石油产品还在中国的出口中充当重要角色。
- 1977年中国的贸易量在世界贸易中的比重为0:6

**贸易自由化**

- 政府逐步停止对进出口的直接规划，更多地依赖于其他贸易政策的复杂组合，包括像关税和配额以及限制贸易权限和更严格的商品检查要求等工具
- 改革后，获得贸易权限的国内企业的数量在快速和大量增加，从最初的12家扩大到1985年的800家左右。到2001年，这一数字扩大到35000家

**出口加工体制**

- 出口加工企业的特权制度于1979年设立。而到了1987年，政府对用于生产出口商品的所有进口的原材料、零部件实施免税。
- 合资企业和外商独资公司也大都被允许进口免税的资本商品

**90年代的保护主义**

- 到80年代中期，中国一直实行的是两种贸易体制，即对外国公司和国内从事出口加工的企业实行非常开放的制度，而对其他企业实行比较限制性的贸易制度。在90年代后半期，中国贸易体制的“双轨性”在下降。
- 大量免税进口的商品在国内市场上非法销售。

**外汇改革**

- 在改革前，为了补助国内无法生产的资本品的进口，汇率一直被高估。
- 在改革中，政府逐步降低了官方的汇率，从1985年的1:5元人民币兑1美元到1994年的8:7元人民币兑1美元。
- 除了大幅度的实际贬值外，中国的出口商也被允许保留部分外汇收入，个人可以持有外汇，资本流出的限制也被放松了。

**外商直接投资的开放**

- 改革后，省和地方政府被赋予相当大的自由权，可以管理在其辖区内所设立的合资企业。对于经济开放区，给予类似税收和行政手续的免除。
- 1986年，FDI法规的变动代表了全国范围的一个大的开放过程。不论地域，外资企业都享有较低的企业所得税率的资格，并且被给予更大的经营自主权。

**中国的出口结构**

- 15年前，中国的主要出口品是原油、加工石油产品和服装。
- 近年来，中国成为了像消费类电子产品、办公设备和计算机以及通讯设备等电子和信息产品的主要生产者和出口者。中国之所以能够出口大量的电子和信息技术产品，只是因为中国进口了大部分出口产品的高增加值的零部件。
