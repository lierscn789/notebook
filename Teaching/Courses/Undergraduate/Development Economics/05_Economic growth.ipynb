## 新古典增长模型——索洛模型

>所有的理论都建立在并不现实的假设之上，这是它成为理论的原因。——索洛

### 总量生产函数

如果$Y$代表总产出，$K$是资本存量，$L$是劳动供给，那么总量生产函数可以表述为如下形式：

$$Y = F(K, L)$$

新古典生产函数中的资本-劳动比不再是固定的，而是根据资本和劳动的相对禀赋，以及生产过程而变化。如果生产函数方程两边同时除以$L$,于是得到

$$Y/L = F(K/L,1)$$

可以简单表述为：

$$y = f(k)$$

索洛模型假设，生产函数具有熟悉的资本收益递减的性质。

<br>

### 资本积累方程

资本存量的变换（$\Delta K$）等于储蓄（$sY$）减去折旧（$\delta K$）

$$\Delta K=s Y-\delta K$$

人均形式

$$\Delta k=s y-(n+\delta) k$$

人均资本的变化（$\Delta k$）取决于三个方面：

- $\Delta k$与人均储蓄（投资）正相关
- $\Delta k$与人口增长负相关
- 折旧使资本存量降低

经济增加人均资本存量$k$的过程称为资本深化。资本存量的增长与劳动力增长和折旧保持同步，被称为资本广化。

<br>

### 索洛模型的稳态

<img src="./pic/solow01.png" width = 50% height = 50% />

A点是新储蓄额$sy$等于新增工人和折旧所需新增资本的唯一点。在该点，每个工人的资本量保持不变了。因此A点称为索洛模型的稳态。稳态的人均产出也被称为每个工人的稳态产出水平、长期产出水平或潜在人均产出水平。

### 经济增长的源泉

索洛开创了基于实际数据基础上的增长的源泉分析框架，对增加产出的因素——资本和劳动的积累、技术进步对经济增长的贡献进行了分析。资本和劳动的贡献在技术上是可以直接测算的，而技术进步对增长的贡献是通过某种“残余”被计算出来的。因此，技术进步的贡献称为“索洛余值”，或被描述为全要素生产率（TFP）。索洛生产函数转化为各要素贡献程度的形式

$$g_{Y}=\left(W_{k} \times g_{K}\right)+\left(W_{L} \times g_{L}\right)+\alpha$$

其中$W_{k}$和$W_{L}$分别代表总收入中工资和资本报酬的比重，$g_{Y}$、$g_{K}$与$g_{L}$分别代表经济、资本存量与劳动力的增长率，残差$\alpha$即全要素生产率。

## 技术进步

### 概念

所谓技术进步，就是使用相同的收入要素组合生产更多的产出，或者反过来，使用更少的投入要素组合来生产同样多的产出。

### 技术进步的分类

- 中性技术进步—— 在资本和劳动力的相对价格不变的情况下，不改变资本和劳动投入比例的技术进步， 例如管理经验的提高，制度的改进
- 非中性技术进步（有偏的技术进步）

**资本偏向的技术进步**

在给定资本和劳动力的相对价格的前提下，资本-劳动比例增加的技术进步。

**劳动偏向的技术进步**

在给定资本和劳动力的相对价格的前提下，资本-劳动比例减少的技术进步。

**增强劳动力的技术进步**

提高劳动力技能的技术进步，类似于“质量-数量”替代。

**增强资本的技术进步**

增强资本的技术进步，例如对土地的改良。

<br>

**诱导性技术进步（变迁）**

技术的选择由要素的相对价格决定；当要素相对价格变化时，人们倾向于采用更密集地使用变得更为便宜的要素的技术。例如，速水佑次郎和弗农拉坦详细考察了1880-1980年间日本和美国农业发展过程中要素价格和技术变迁的关系。美国劳均土地面积远远高于日本,劳动相对于土地的价格也高于日本,但是两国的农业生产和生产率都取得了高速增长。正如诱致性技术创新理论揭示的那样,美国劳动生产率的增长主要依靠机械技术进步带来单位劳动力拥有土地面积的增加,而日本则依靠生物技术进步促进土地生产率的增长。美国和日本的历史经验充分证实农业增长能够被理解为一个由相对价格变动所引起的沿着长期生产函数移动的要素替代过程。

## 经济收敛性的讨论

### $\sigma$收敛与$\beta$收敛

索洛模型的一个重要结论是，在没有技术进步的情况下，稳态上人均收入为常数，即对于人口增长率和储蓄率相同的国家而言，它们的人均收入将收敛到相同的水平上。在文献里，通常把这种人均收入水平的收敛称为$\sigma$收敛，即国家之间人均收入分布的方差缩小。而另一个必然结论是，随着人均收入（资本）的提高，并且越来越接近稳态水平，一个国家的增长速度将下降，在文献中，这个过程被称为$\beta$收敛。

<br>

### 绝对收敛和条件收敛

所谓绝对收敛，指的是无论国家之间的稳态是否相同，都发生收敛；而条件收敛意味着，只有稳态相同的国家之间才能发生收敛。

## 经济增长的模式

经济增长依靠以下两个基本过程

- 要素积累，指资本存量规模或劳动力的增加
- 生产率的提高，主要利用技术进步

快速增长型国家的特征

- 宏观经济和政治上的稳定

    **两个刚果**
    - 在非洲中西部以刚果为名的国家有两个：刚果共和国（简称刚果或刚果布）和刚果民主共和国（简称民主刚果或刚果金）。
    - 刚果民主共和国在1990-2002年间经历了惊人的年均2800%通货膨胀率，每年7.2%的负增长，预期寿命从52岁下降到45岁，婴儿死亡率从每千人128上升到139。


- 对健康和教育的投资

    **马来西亚对减少疟疾和改善健康的努力**
    - 为了寻找锡，当定居者最初于19世纪50年代来到吉隆坡时，有一半人死于疟疾。一个世纪后，情况得到了改善。一项积极的政府疟疾控制计划起到了非常重要的作用。到1975年，疟疾病例数与1960年相比，已经减少了2/3，预期寿命从1960年的54岁增加到64岁。健康的改善（与其他因素一起）促进了增长的涌现，使得1976年——1996年间增长率加速到每年近5%。


- 有效的监管和制度

    **诺斯悖论** 
    - 所谓诺斯悖论，指在经济活动当中，国家提供的基本服务是制定游戏规则，特别是界定产权制度的基本规则。没有国家权力及其代理人的介入，财产权利就无法得到有效保护和实施。另一方面，国家常常会出于私利而建立和维持无效的产权制度，从而导致所有权残缺，造成无效产权和经济的衰落。用诺斯本人的话说，“没有国家办不成事，有了国家又有很多麻烦”。这就是国家的两难，或者诺斯悖论。


- 有利于私有企业的环境

    **谢尔曼反垄断法**
    - 1890年美国国会通过了一项前所未有的法律“谢尔曼反垄断法”，这项法律的主旨是保护自由竞争的市场体系、不允许行业大佬通过垄断手法妨碍自由竞争。这是美国独立以后一百多年来，联邦政府第一次被授权保护经济领域竞争。罗斯福担任总统后得到大力运用，他下令调查、起诉标准石油公司，审理案件旷日持久，直到1911年，最高法院才最终作出了裁决，标准石油公司被下令解散，分为33家公司。
    

- 有利的地理环境

    **地域决定论**
    - 地域决定论认为, 一国或地区的地理位置决定其资源丰裕程度、气候优劣、交通和区位优势, 将直接影响物质资本投资、人力资本质量和技术进步。在控制政策和制度影响后, Sachs( 2001)发现,地域往往通过交易成本、交通卫生状况和农业生产率等途径影响经济产出。诸如沿海地区利用便利的交通条件、低廉的运输成本和参与国际贸易的潜在优势, 更易于快速实现经济增长。而不利区位以及炎热气候将导致热带地区过快的人口增长和过重的医疗负担, 进而抑制经济发展。

## 市场与政府作用

### 市场与市场失灵

观点：大多数经济活动应当由市场来管理，而非通过国家的行政干预。

- 市场能够更好地进行资源分配

- 市场通常比政府更加灵活

- 市场增强了竞争

<br>

但毫无疑问，市场本身可能存在问题，即市场失灵（market failure）

- 垄断或寡头力量的产生

- 外部经济

- 外部不经济

- 市场不能提供发展所需要的经济结构转变 —— 例如幼稚产业

- 不完全信息

### 实行市场改革

- 结构调整

    - 价格调整
    - 保障竞争
    - 产权制度

- 把握市场改革的进度

    - 休克疗法
    - 渐进式改革
