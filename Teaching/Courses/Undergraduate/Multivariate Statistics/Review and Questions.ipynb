OLS estimator for simple regression model

$$\hat{\beta}_{0}=\overline{y}-\hat{\beta}_{1} \overline{x}$$

$$\hat{\beta}_{1}=\frac{\sum_{i=1}^{n}\left(x_{i}-\overline{x}\right)\left(y_{i}-\overline{y}\right)}{\sum_{i=1}^{n}\left(x_{i}-\overline{x}\right)^{2}}$$

The R-squared of the regression, sometimes called the coefficient of determination, is defined as

$$R^{2} \equiv \mathrm{SSE} / \mathrm{SST}=1-\mathrm{SSR} / \mathrm{SST}$$

$R^{2}$ is the ratio of the explained variation compared to the total variation; thus, it is interpreted as the fraction of the sample variation in y that is explained by x. 

t test

Generally, if the null is stated as

$$\mathrm{H}_{0} : \beta_{j}=a_{j}$$

then the appropriate t statistic is

$$t=\left(\hat{\beta}_{j}-a_{j}\right) / \operatorname{se}\left(\hat{\beta}_{j}\right)$$

The general t statistic is usefully written as

$$t=\frac{(\text {estimate - hypothesized value})}{\text {standard error}}$$

Testing Multiple Linear Restrictions: The F Test

$$F \equiv \frac{\left(\mathrm{SSR}_{r}-\mathrm{SSR}_{u r}\right) / q}{\operatorname{SSR}_{u r} /(n-k-1)}$$

or

$$F=\frac{\left(R_{u r}^{2}-R_{r}^{2}\right) / q}{\left(1-R_{u r}^{2}\right) /(n-k-1)}=\frac{\left(R_{u r}^{2}-R_{r}^{2}\right) / q}{\left(1-R_{u r}^{2}\right) / d f_{u r}}$$
