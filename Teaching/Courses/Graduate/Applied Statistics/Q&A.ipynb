**习题**

两家影院竞争1000人的顾客源，假设每名顾客对两家影院并无好恶之分，且各人的选择相互独立，令$N$记每家影院的座位数。请问，为了保证因影院客满致使顾客离开的概率小于$1\%$，$N$应为多少？

A：

考虑二项分布

$$P(X=x | n, p)=\left(\begin{array}{l}{n} \\ {x}\end{array}\right) p^{x}(1-p)^{n-x}, \quad x=0,1,2, \cdots, n$$

因此得到

$$P(X>N)=\sum_{x=N+1}^{1000}\left(\begin{array}{c}{1000} \\ {x}\end{array}\right)\left(\frac{1}{2}\right)^{x}\left(1-\frac{1}{2}\right)^{1000-x}<.01$$

求得$N = 537$。

考虑用正态分布近似求值，因此$\mu=1000\left(\frac{1}{2}\right)=500$，$\sigma^{2}=1000\left(\frac{1}{2}\right)\left(\frac{1}{2}\right)=250$，即$X \sim N(500,250)$。因此

$$P(X>N)=P\left(\frac{X-500}{\sqrt{250}}>\frac{N-500}{\sqrt{250}}\right)<.01$$

即

$$P\left(Z>\frac{N-500}{\sqrt{250}}\right)<.01$$

查表得到

$$\begin{aligned} P(Z>2.33) \approx .0099<.01 & \Rightarrow \frac{N-500}{\sqrt{250}}=2.33 \\ & \Rightarrow N \approx 537 \end{aligned}$$

**习题**

设$X_{1}, \cdots, X_{n}$是iid的$N\left(\theta, \sigma^{2}\right)$样本，求$\theta$和$\sigma^{2}$的矩法估计量。

A：

根据

$$\begin{aligned} \operatorname{Var} X &=\mathrm{E}(X-\mathrm{E} X)^{2}=\mathrm{E}\left[X^{2}-2 X \mathrm{E} X+(\mathrm{E} X)^{2}\right] \\ &=\mathrm{E} X^{2}-2(\mathrm{E} X)^{2}+(\mathrm{E} X)^{2} \\ &=\mathrm{E} X^{2}-(\mathrm{E} X)^{2} \end{aligned}$$

我们有

$$m_{1}=\overline{X}, m_{2}=(1 / n) \sum X_{i}^{2}, \mu_{1}^{\prime}=\theta, \mu_{2}^{\prime}=\theta^{2}+\sigma^{2}$$

因此，求解

$$\overline{X}=\theta, \frac{1}{n} \sum X_{i}^{2}=\theta^{2}+\sigma^{2}$$

得到矩法估计量

$$\tilde{\theta}=\overline{X}$$

$$\tilde{\sigma}^{2}=\frac{1}{n} \sum X_{i}^{2}-\overline{X}^{2}=\frac{1}{n} \sum\left(X_{i}-\overline{X}\right)^{2}$$

**习题**

假设下列数字来自于正态分布的总体：

90.46561，105.1319，117.5445，102.7179，102.7788，107.6234，94.87266，95.48918，75.63886，87.40594

假设这些数字是独立的，求$\mu$和$\sigma^{2}$的MLE？

A：

考虑到这些数字的独立性，以及它们符合正态分布，因此得到概率密度函数的乘积

$$L\left(\mu, \sigma^{2}\right)=\prod_{i=1}^{n} \frac{1}{\sqrt{2 \pi \sigma^{2}}} \exp \left(\frac{-\left(y_{i}-\mu\right)^{2}}{2 \sigma^{2}}\right)$$、

得到似然函数

$$L\left(\mu, \sigma^{2}\right)=\left(\frac{1}{\sqrt{2 \pi \sigma^{2}}}\right)^{n} \exp \left(-\frac{1}{2 \sigma^{2}} \sum_{i=1}^{n}\left(y_{i}-\mu\right)^{2}\right)$$

取对数，得到对数似然函数

$$\ln L\left(\mu, \sigma^{2}\right)=-\frac{n}{2} \ln \left(2 \pi \sigma^{2}\right)-\frac{1}{2 \sigma^{2}} \sum_{i=1}^{n}\left(y_{i}-\mu\right)^{2}$$

通过对数似然函数对两个未知参数$\mu$和$\sigma^{2}$求偏导设置为零，对于$\mu$

$$\frac{\partial \ln \left(L\left(\mu, \sigma^{2}\right)\right)}{\partial \mu}=2 \frac{1}{2 \sigma^{2}} \sum_{i=1}^{n}\left(y_{i}-\mu\right)=0$$

上述表达式可以简化为

$$\sum_{i=1}^{n}\left(y_{i}-\hat{\mu}\right)=0$$

因此得到$\hat{\mu}=\frac{\sum_{i=1}^{n} y_{i}}{n}=\overline{y}$。同样，如果对于$\sigma^{2}$求偏导为零，可以最终得到

$$\hat{\sigma}^{2}=\frac{\sum_{i=1}^{n}\left(y_{i}-\hat{\mu}\right)^{2}}{n}$$

可以看到它是有偏的，我们应该除以$(n-1)$替代。

**习题**

欧文草坪护理有限公司生产和组装割草机。现在存在下列两个不同的发动机装配到割草机的框架工序的提议，问题是两个工序花费的平均时间是否不同，即平均组装时间是否有差异？

$$\begin{array}{ll}{\text { Welles }} & {\text { Atkins }} \\ \hline \text { (minutes) } & {\text { (minutes) }} \\ \hline 2 & {3} \\ {4} & {7} \\ {9} & {5} \\ {3} & {8} \\ {2} & {4} \\  & {3}\end{array}$$

A:

计算样本均值和标准差

$$\bar{x}_{W}=\frac{\Sigma x_{W}}{n_{W}}=\frac{20}{5}=4 \quad \bar{x}_{A}=\frac{\Sigma x_{A}}{n_{A}}=\frac{30}{6}=5$$

$$s_{W}=\sqrt{\frac{\Sigma\left(x_{W}-\bar{X}_{W}\right)^{2}}{n_{W}-1}}=\sqrt{\frac{34}{5-1}}=2.9155 \quad s_{A}=\sqrt{\frac{\Sigma\left(x_{A}-\bar{X}_{A}\right)^{2}}{n_{A}-1}}=\sqrt{\frac{22}{6-1}}=2.0976$$

计算混合样本的方差

$$s_{p}^{2}=\frac{\left(n_{w}-1\right) s_{w}^{2}+\left(n_{A}-1\right) s_{A}^{2}}{n_{w}+n_{A}-2}=\frac{(5-1)(2.9155)^{2}+(6-1)(2.0976)^{2}}{5+6-2}=6.2222$$

得到$t$统计量

$$t=\frac{\bar{x}_{W}-\bar{x}_{A}}{\sqrt{s_{p}^{2}\left(\frac{1}{n_{W}}+\frac{1}{n_{A}}\right)}}=\frac{4.00-5.00}{\sqrt{6.2222\left(\frac{1}{5}+\frac{1}{6}\right)}}=-0.662$$

结论是不能拒绝原假设
