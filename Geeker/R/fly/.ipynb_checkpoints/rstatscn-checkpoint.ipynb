```r
statscnDbs()
```

```r
category <- as_tibble(statscnQueryZb(zbid = "zb", dbcode = "fsnd"))

category

dplyr::filter(category, isParent==TRUE)
```

```r
# 获取所有变量及名称
variables <- function(category){
    vars <- tibble()
    df1 <- as_tibble(statscnQueryZb(zbid = category, dbcode = "fsnd"))
    vars <- rbind(vars, df1 %>% dplyr::filter(isParent==FALSE) %>% select(id, name))
    df2 <-  df1 %>% dplyr::filter(isParent==TRUE)
    if (dim(df2)[1] > 0) {
        for (i in 1:dim(df2)[1]){
            vars <- rbind(vars, variables(df2$id[[i]]))
        }
    }
    vars
}

variables(category="A0S")
```
